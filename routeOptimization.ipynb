```python
import osmnx as ox
import networkx as nx
import folium


G = ox.graph_from_place('Baku, Azerbaijan', network_type='drive')


start_point = (40.4093, 49.8671)  # Örneğin Baku'da bir yer
end_point = (40.3877, 49.8822)    # Baku'da başka bir yer

start_node = ox.distance.nearest_nodes(G, X=start_point[1], Y=start_point[0])
end_node = ox.distance.nearest_nodes(G, X=end_point[1], Y=end_point[0])


shortest_path = nx.shortest_path(G, start_node, end_node, weight='length')


route_coords = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in shortest_path]


m = folium.Map(location=start_point, zoom_start=13)


folium.Marker(location=start_point, popup='Start Point', icon=folium.Icon(color='green')).add_to(m)
folium.Marker(location=end_point, popup='End Point', icon=folium.Icon(color='red')).add_to(m)


folium.PolyLine(locations=route_coords, color='blue', weight=5, opacity=0.7).add_to(m)


m.save('shortest_route.html')

# Alternatif olarak, Jupyter Notebook'ta görüntülemek için:
m
```
