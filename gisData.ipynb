```python
import requests
import polyline  # Make sure you have installed this package: `pip install polyline`
from gtts import gTTS
import os
from pydub import AudioSegment  # Make sure you install pydub and ffmpeg

# Replace this with your actual API key from GraphHopper
api_key = 'apikey'

# Coordinates for the start and end points
start_coords = '40.374899,49.85693'  # Start point
end_coords = '40.376467,49.852033'  # End point

# GraphHopper API endpoint
url = "https://graphhopper.com/api/1/route"
params = {
    'point': [start_coords, end_coords],  # Use your actual start and end coordinates
    'vehicle': 'bike',
    'locale': 'en',
    'instructions': 'true',
    'key': api_key
}

# Make the request to GraphHopper API
response = requests.get(url, params=params)
data = response.json()

# Check if the API request was successful
if 'paths' in data and len(data['paths']) > 0:
    # Extract the path (route) from the response
    path = data['paths'][0]
    distance = round(path['distance'])  # Rounded distance to nearest meter
    time = round(path['time'] / 1000)  # Convert milliseconds to seconds and round

    encoded_polyline = path['points']

    # Decode the polyline
    decoded_points = polyline.decode(encoded_polyline)

    # Extract turn-by-turn instructions
    instructions = path['instructions']

    # Output the data
    print(f"Total Distance: {distance} meters")
    print(f"Total Time: {time} seconds")
    print("Route Coordinates:")
    for point in decoded_points:
        print(point)

    print("\nTurn-by-Turn Instructions:")
    full_text_instructions = ""
    for instruction in instructions:
        instruction_distance = round(instruction['distance'])  # Rounded instruction distance
        # Replace "küçəsi" with "street" in each instruction
        text = instruction['text'].replace("küçəsi", "street")
        text += f" for {instruction_distance} meters"
        print(text)
        full_text_instructions += text + ". "

    # Split the instructions into sentences
    sentences = full_text_instructions.split(". ")

    # Create a blank audio segment for delay
    silence = AudioSegment.silent(duration=2000)  # 2 seconds of silence

    # Combine sentences with a delay between them
    final_audio = AudioSegment.empty()
    for sentence in sentences:
        if sentence.strip():  # Check if the sentence is not empty
            tts = gTTS(sentence, lang='en')
            tts.save("temp.mp3")
            sentence_audio = AudioSegment.from_mp3("temp.mp3")
            final_audio += sentence_audio + silence

    # Save the final audio file with pauses
    final_audio.export("instructions_with_delay.mp3", format="mp3")
    print("Turn-by-turn instructions with delay have been saved to 'instructions_with_delay.mp3'.")

    # Optional: Play the mp3 file
    os.system("start instructions_with_delay.mp3")  # For Windows users
    # For MacOS or Linux use: os.system("open instructions_with_delay.mp3") or os.system("mpg123 instructions_with_delay.mp3")
else:
    print("No paths found in the API response. Please check your inputs or the API key.")


```
